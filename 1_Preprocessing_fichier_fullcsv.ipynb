```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("full.csv", sep=',', encoding = 'UTF-8')
df.head()
```

```python
# Pour connaitre le nombre de lignes et colonne de df
df.shape
```

### 1-Selection des colonnes qu'on veut garder

```python
df2=df[['id_parcelle',
        'nature_mutation',
        'date_mutation',
        'valeur_fonciere', 
        'nombre_lots', 
        'lot1_surface_carrez',
        'surface_reelle_bati',
        'type_local',
        'nombre_pieces_principales',
        'code_departement',
        'adresse_numero',
        'adresse_nom_voie', 
        'code_postal',
        'longitude', 
        'latitude']]
df2.head()
```

```python
# Pour connaitre le nombre de lignes et colonnes de df2
df2.shape
```

```python
# Selectionner que les élements dans la colonne type_local qui correspond à "Appartement" 
# Car choix de ne regarder que les appartements pour le projet
df2 = df2[df2['type_local']=="Appartement"]
df2.head()
```

```python
# Verifier le nombre de lignes restantes
df2.shape  # il reste 148646 lignes
```

```python
# Selectionner que les appartements situés dans la région parisienne 
df2=df2[df2['code_departement'].astype(str).str.contains('75|78|91|92|93|94|95')]
df2.head(2)
```

```python
# Verifier le nombre de lignes restantes après le trie sur la région parisienne
df2.shape  # il reste 48170 lignes
```

```python
# Vérification des données manquantes dans le dataframe trié df2
df2.isnull().sum() # ce code permet de faire la somme du nombre 'NaN' par colonne
```

```python
# Pour voir un échantillon au hasard de 5 lignes
df2.sample(5)
```

```python
# Enlever la colonne "lot1_surface_carrez" pour plus de clarté dans le dataframe df2 trié
df2.drop(columns= ['lot1_surface_carrez'], inplace=True)
df2.sample(3)
```

```python
# Pour garder les lignes de "valeur_foncière" renseignées
# C'est à dire, enlever les 77 valeurs manquantes de la colonne valeur_fonciere vue précedement
df2 = df2[df2['valeur_fonciere'].notna()]
df2.sample(3)
```

```python
df2.shape # on retrouve bien (48170 - 77 => Soit 48093 lignes)
```

```python
df2.columns
```

```python
# Pour connaitre les différentes valeurs dans la colonne "nature_mutation"
df2[df2['nature_vente']=='Vente'].count()
```

```python
# Trier maintenant dans la colonne "nature_mutation" uniquement les valeurs "Vente" 
df2[(df2['nature_mutation']=='Vente')]
# Attention ce code prend toutes les ventes, il reste les 2 ventes de terrain à bâtir que je n'arrive pas à virer.
df2.shape #(47698 + 211 + 2 =47911 lignes)
```

```python
# Réunir les colonnes adresse_numero, adresse_nom_voie et code_postal dans une nouvelle colonne "Adresse"
df2["Adresse"] = df2["adresse_numero"].astype(str) + "   "+ df2["adresse_nom_voie"]+ "   " + df2["code_postal"].astype(str)
df2.head()
```

```python
# Emlever la colonne adresse_numero
df2.drop(columns= ['adresse_numero'], inplace=True)
df2.sample(3)
```

```python
# Renommer les colonnes pour plus de clarté

df2.rename(columns={'nature_mutation': 'nature_vente', 
                   'date_mutation': 'date_vente',
                   'valeur_fonciere': 'Prix',
                   'nombre_lots': 'nb_lots',
                   'surface_reelle_bati': 'surface',
                   'type_local': 'local',
                   'nombre_pieces_principales': 'nb_pieces',
                   'code_departement': 'code_dprtmt'               
                  }, inplace=True)
df2.head()
```
